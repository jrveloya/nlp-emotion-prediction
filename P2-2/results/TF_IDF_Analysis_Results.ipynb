# (Term Frequency-Inverse Document Frequency) TF-IDF Methodology

Term Frequency-Inverse Document Frequency (TF-IDF) is a frequency-based natural language processing method used to evaluate the importance of a word in a text based on a collection of texts. It consists of two main components, term frequency (TF) and inverse document frequency (IDF). The TF portion computes how frequently a word appears on a text, while the IDF portion increases the weight of less frequent words and decreases the weight of common words (ie: stop words) that appear in multiple texts, which means a word that appears in fewer texts is likely to provide more meaningful context. With these two components, TF-IDF is able to track which words are the most frequent in a specific text and are unique throughout a collection of texts.

The text is first set to lowercase and tokenized. If the option to remove stopwords and/or to apply stemming via the Porter Stemmer algorithm has been specified, then words will have their stems removed, as well as removing stop words from the tokens. TF-IDF then vectorizes the words, converting them into numerical vectors as n-grams. The vectorizer then learns the vocabulary and weighting of the text and scales these vectors. The TF-IDF scores are then computed using the cosine similarity metric, which is then compared to the human-based similarity matrix to compute the Pearson correlation between the matrices.

# TF-IDF Analysis and Results

There are 12 different configurations setup for TF-IDF. Two options to either keep or remove stop words, two options to either perform stemming or not, and three options to vectorize into unigrams, bigrams, or trigrams. Performance on the unigram configurations with no stop words resulted in a Pearson correlation of 0.5513 and 0.5347 for no stemming and stemming, respectively. Bigram and trigram configurations with no stopwords have shown progressively worse performance, with Pearson correlations of 0.4456 (No stemming) and 0.4454 (Stemming) for bigrams and 0.3833 (No stemming) and 0.3880 (Stemming) for trigrams, respectively. Keeping stopwords degrades unigram configuration performance with Pearson correlations of 0.4941 (No stemming) and 0.4865 (Stemming), while bigram and trigram configurations has shown improved correlation to the human-based similarity matrix, yielding 0.4858 (No stemming) and 0.4801 (Stemming) for bigrams and 0.4209 (No stemming) and 0.4225 (Stemming) for trigrams, respectively. The improved correlation of bigram and trigram configurations when keeping stop words is probably caused by generating bigrams and trigrams on multi-word expressions, retaining the context of the expressions, whereas with unigrams, the context of these expressions is lost. Overall, bigram and trigram configurations' lower performance in comparison with unigram configurations is likely caused by data sparsity of the vocabulary, resulting in a low frequency of many n-gram tokens and the overfitting of the vectorizer. Additionally, stemming in general, decreases correlation, likely caused by crudely reducing words to their base form, removing semantic context and nuance of the words.
