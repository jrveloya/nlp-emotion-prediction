# Pearson Analysis and Methodology

After completing these steps, we now have 28 NLP similarity matrices along with 1 human-based similarity matrix. In order to evaluate which methods match human perception the best, we compare each of these NLP similarity matrices with the human based matrix. To accomplish this, we extracted half of each NLP matrix, and compared each of those halves with half of the human matrix. These halves were obtained by taking the upper triangle of each matrix, while excluding the diagonal. We used NumPy's triu_indices() to do this. After extracting the halves, we calculated the Pearson Correlation coefficient for all 28 NLP similarity matrices. Pearson Correlation in our case calculates the linear relationship between two matrices. Scores range from -1.0 to 1.0, where -1.0 represent a perfect inverse linear relationship, while 1.0 represents a symmetrical linear relationship. (Insert formula here, or make a figure for it). 

# Pearson Results

Results from the comparisons show our best techniques receiving a pearson correlation coefficient around 0.556. **The best performing technique overall used WordNet with stopwords removed and no lemmatization used and received a coefficient number of 0.556867**. The best configuration for Bag of Words included no stopwords, no stemming, and used unigrams. **This configuration for Bag of Words achieved a coefficient number of 0.556294**. The best configuration for TF-IDF included no stopwords, no stemming, and used unigrams. **This configuration for TF-IDF achieved a coefficient number of 0.551306**. Considering these results, it seems that removing stopwords has a positive effect on the correlation score, as most of the top performing configurations included the removal of stopwords. It also seems as if stemming and lemmatization have little to no effect on the correlation score. This may be due to the original corpus having a simple vocabulary with mostly root-level words. For Bag of Words and TF-IDF, unigram achieved the overall best coefficient scores. Intuitively, unigram appears to perform the best due to the short and single sentence entries in the corpus. With these relatively small entries, using unigrams and comparing one word at a time makes the most sense. Comparatively, using bigrams and trigrams is less efficient, due to the difficulty of findings two and three word trends in these smaller entries. When comparing among the three techniques, WordNet, Bag of Words, and TF-IDF, they all appear to be very similar when comparing to the human matrix. Their highest pearson coefficient numbers are all within approximately 0.01 of each other, indicating very minimal difference. Based on this, it seems as if all three techniques are similarly effective for this particular corpus.
