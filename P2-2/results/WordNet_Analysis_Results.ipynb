# WordNet Methodology

WordNet is a lexical database, primarily for the English language, that links words, such as nouns, into semantic relations. These semantic relations are grouped and organized into synsets, which are a set of synonyms sharing a common meaning. This makes WordNet function similarly to a thesaurus. 

To prepare the text for synset extractions, part-of-speech tagging is performed on the text to apply Treebank tags to the words. Then these Treebank tags are mapped into WordBank tags. If the tag isn't a noun, verb, adjective, or adverb, then the tag is discarded. Stopwords are either removed or kept depending on the configuration settings. Likewise, when looking for the lookup form of a word, the lemma or the text of the word is extracted. Once we extracted the lookup form and got the mapped WordBank tag, we then look up all possible synsets of the target word and get the most frequent synset to append to the list of synsets. 

Once the most frequent synsets of every sentence is extracted, the synsets compute to each other. If the synsets are identical (i = j), then we compute 1.0 for that cell in the matrix. Otherwise, we will compute the similarity score using Greedy Lemma Alignment Overlap (GLAO). For every lemma/word in the smaller synset, we attempt to greedily find the best similarity score in respect to the larger synset while keeping tracking of the index in the larger synset. Once the best similarity score for the lemma/word pair has been obtained, we add to the summation of similarity scores and don't consider this pair in the computation afterward. Once the summation of similarity scores has been obtained, the summation is divided by the length of the larger synset to obtain the mean of the similarity scores. Afterward, the computation is repeated in inverse order, added to the original computation for symmetry, as the sentences are not equal to each other. It is then divided by 2 to obtain the mean symmetric sentence similarity.  

# WordNet Results and Analysis

Four different configurations were tested, all of which were computed using GLAO and symmetric sentence similarity to obtain the similarity matrix. To get the Pearson Correlation, the upper triangle of the human-based similarity matrix and the Wordnet-based similarity matrices are extracted. These triangles are then used to compute the Pearson correlation of the Wordnet-based similarity matrices. The no stopwords and no lemmatization configurations obtained a Pearson correlation of 0.55786, followed by the configuration with no stopwords, but with lemmatization, reaching a Pearson correlation of 0.55657. However, keeping stopwords degrades the Pearson correlation as both configurations that included stopwords yielded a Pearson correlation of 0.52924 and 0.52923, with and without lemmatization respectively. The configurations with no stopwords outperform all BOW and TF-IDF vectorizer configurations, while the configurations with stopwords outperform all vectorizer configurations except the BOW and TF-IDF vectorizer configurations with no stopwords and using unigrams. This appears to indicate that WordNet-based similarity matrices correlate more than the BOW and TF-IDF methods. However, these results seem to contradict the results Abri et al. reported, obtaining 0.425 (No lemmatization) and 0.373 (Lemmatization) Pearson correlation in the configurations with no stopwords and 0.366 (No lemmatization) and 0.299 (Lemmatization) Pearson correlation in the configurations with the stopwords kept. It is likely that obtaining only the most frequent synsets for every word in the text has significantly improved the correlation between the human-based matrix and the WordNet-based matrices, given that Abri et al. mentioned that correlations between the human-based matrix and frequency-based approaches resulted in higher correlations.
