# MLP Methodology

A multi-layer perceptron (MLP) is a neural network model consisting of an input layer, multiple hidden layers, and an output layer. The input data is fed to the input layer and through all the hidden layers. Each layer performs computations based on the input data with an activation function to add nonlinearity, allowing the network to learn patterns in the data. A loss is computated using a binary cross-entropy loss function, which is used in backpropagation to update the MLP's weights and bias via gradient descent to minimize the loss function. Once all hidden layers have been passed through, the data from the final hidden layer is fed to the output layer, which outputs the final prediction.

The glove-wiki-giga-100 GloVe pre-trained model was used to perform word embeddings on the text. The data is first preprocessed by tokenizes all input, performing lemmatization, and part of speech tagging. All stopwords, punctuation, as well as lemmas that aren't in the GloVe model's vocabulary are omitted. The vector embeddings of all valid lemmas in the input texts are computed to get the mean of the text vector embeddings. These mean vector embeddings are scaled using the StandardScaler from sklearn and Principal Component Analysis (PCA) is performed to reduce the vector embeddings to the top 20 embeddings to mitigate overfitting. 5-fold cross-validation and hyperparameter tuning are performed via grid search with the training set to search the best possible hyperparameters for the classifier. The best model is saved using the joblib library and the training and test classification reports, accuracy, and f1 scores are printed.

# Results Analysis

The MLP classifier model was trained on a stratified-split dataset of 80:20, resulting in approximately a 50-50 distribution between the positive and negative sentiment samples for all subsets. The best model parameters were set to the following parameters: activation: 'logistic', alpha: 10.0, early_stopping: True, hidden_layer_size: 8, n_iter_no_change: 10, solver: 'lbfgs'. For training results, the classifier model achieved 0.9375 accuracy and 0.9374 f1-score and in test, the model achieved 0.8077 and 0.8065 f1-score, indicating the overfitting of the model. Possible ways to reduce overfitting is by employing the use of Sentence-Bidirectional Encorder Representation from Transformers (SBERT), which utilizes whole sentences to incorporate context, allowing it to easily capture word meanings and sentence structures and to use the Valence Aware Dictionary sEntiment Reasoner (VADER) tool to calculate the polarity and intensity of sentiments in the text.
